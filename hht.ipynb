```python
import numpy as np
import librosa
import librosa.display
from scipy.signal import hilbert
import matplotlib.pyplot as plt
```

```python
# Carregue o arquivo MP3
mp3_file = 'audio1.mp3'
y, sr = librosa.load(mp3_file)

analytic_signal = hilbert(y)

# Crie o vetor de tempo
t = np.arange(0, len(y)) / sr

# Plote o sinal de áudio
plt.figure(figsize=(12, 6))
plt.subplot(2, 1, 1)
plt.plot(t, y)
plt.title('Sinal de Áudio')
plt.xlabel('Tempo (s)')
plt.ylabel('Amplitude')

# Plote a parte real da Transformada de Hilbert
plt.subplot(2, 1, 2)
plt.plot(t, np.real(analytic_signal))
plt.title('Parte Real da Transformada de Hilbert')
plt.xlabel('Tempo (s)')
plt.ylabel('Amplitude')

plt.tight_layout()
plt.show()
```

```python
import numpy as np
import matplotlib.pyplot as plt
from scipy.io import wavfile
from pyhht.emd import EMD
from pyhht.visualization import plot_imfs

# Carregue o arquivo de áudio (substitua 'audio.wav' pelo caminho do seu arquivo)
sample_rate, audio_data = wavfile.read('audio.wav')

# Normalize o áudio para valores entre -1 e 1
audio_data = audio_data.astype(float) / 32767.0

# Crie um vetor de tempo com base na taxa de amostragem e duração do áudio
time = np.arange(0, len(audio_data)) / sample_rate

# Crie um objeto EMD e decomponha o áudio em IMFs
decomposer = EMD(audio_data)
imfs = decomposer.decompose()

# Plote as IMFs
plot_imfs(audio_data, imfs, time)

# Mostre o gráfico
plt.title("Decomposição HHT do Áudio")
plt.xlabel("Tempo (s)")
plt.ylabel("Amplitude")
plt.show()
```
